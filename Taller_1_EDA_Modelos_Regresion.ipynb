```python
### iMPORTAR DEPENDENCIAS ###
import pandas as pd
import numpy as np
import pyarrow
import fastparquet
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from statsmodels.graphics.gofplots import ProbPlot
```

```python
    # ==========================
    # CONFIGURACI√ìN
    # ==========================
    PATH = "listings.csv.gz"
    OUT_CSV = "listings_clean.csv"

    MISSING_COL_DROP_THRESHOLD = 0.60
    CRITICAL_COLS = ["id", "price", "latitude", "longitude"]
    GROUP_FOR_IMPUTE = "neighbourhood_cleansed"

    # ==========================
    # 1) CARGA
    # ==========================
    df = pd.read_csv(PATH, compression="gzip", low_memory=False)

    print("\n================ INICIO PROCESO =================")
    print("Shape inicial:", df.shape)
    print("\nPrimeras filas:")
    print(df.head(3))

    # ==========================
    # 2) DIAGN√ìSTICO INICIAL
    # ==========================
    dup_rows = df.duplicated().sum()
    print(f"\nDuplicados completos detectados: {dup_rows}")

    if "id" in df.columns:
        dup_id = df.duplicated(subset=["id"]).sum()
        print(f"Duplicados por ID detectados: {dup_id}")

    missing_ratio = df.isna().mean()
    missing_count = df.isna().sum()

    # ==========================
    # 3) ELIMINAR COLUMNAS CON MUCHOS FALTANTES
    # ==========================
    cols_to_drop = missing_ratio[missing_ratio > MISSING_COL_DROP_THRESHOLD].index.tolist()

    if cols_to_drop:
        print("\nColumnas eliminadas por exceso de faltantes (>60%):")
        for col in cols_to_drop:
            print(f" - {col} | Faltantes: {missing_count[col]} ({round(missing_ratio[col]*100,2)}%)")
        df = df.drop(columns=cols_to_drop)
    else:
        print("\nNo se eliminaron columnas por faltantes.")

    # ==========================
    # 4) LIMPIEZA DE FORMATOS
    # ==========================
    print("\nCorrigiendo formatos...")

    date_cols = ["last_scraped", "calendar_last_scraped", "host_since",
                "first_review", "last_review"]

    for c in date_cols:
        if c in df.columns:
            before_na = df[c].isna().sum()
            df[c] = pd.to_datetime(df[c], errors="coerce")
            after_na = df[c].isna().sum()
            print(f" - Fecha convertida: {c} | Nuevos NaT generados: {after_na - before_na}")

    if "price" in df.columns:
        before_na = df["price"].isna().sum()
        df["price"] = (
            df["price"]
            .astype("string")
            .str.replace(r"[\$,]", "", regex=True)
            .str.strip()
            .replace({"": pd.NA})
        )
        df["price"] = pd.to_numeric(df["price"], errors="coerce")
        after_na = df["price"].isna().sum()
        print(f" - Price convertido a num√©rico | Nuevos NA: {after_na - before_na}")

    # ==========================
    # 5) ELIMINAR DUPLICADOS
    # ==========================
    before_dup = len(df)

    if "id" in df.columns:
        df = df.drop_duplicates(subset=["id"], keep="last")
    else:
        df = df.drop_duplicates()

    after_dup = len(df)
    print(f"\nFilas eliminadas por duplicados: {before_dup - after_dup}")

    # ==========================
    # 6) ELIMINAR FILAS CR√çTICAS
    # ==========================
    critical_existing = [c for c in CRITICAL_COLS if c in df.columns]

    if critical_existing:
        before_rows = len(df)
        df = df.dropna(subset=critical_existing)
        after_rows = len(df)

        print("\nFilas eliminadas por faltantes en columnas cr√≠ticas:")
        print("Columnas cr√≠ticas consideradas:", critical_existing)
        print("Filas eliminadas:", before_rows - after_rows)

    # ==========================
    # 7) IMPUTACI√ìN CONTROLADA
    # ==========================
    print("\nIniciando imputaci√≥n...")

    num_cols = df.select_dtypes(include=["number"]).columns
    cat_cols = df.select_dtypes(include=["object", "string"]).columns
    bool_cols = df.select_dtypes(include=["boolean"]).columns

    def report_impute(col, before, after):
        imputados = after - before
        if imputados > 0:
            print(f" - {col}: {imputados} valores imputados")

    # Num√©ricas
    for c in num_cols:
        before_na = df[c].isna().sum()
        med = df[c].median()
        df[c] = df[c].fillna(med)
        after_na = df[c].isna().sum()
        report_impute(c, after_na, before_na)

    # Categ√≥ricas
    for c in cat_cols:
        before_na = df[c].isna().sum()
        mode = df[c].mode()
        if len(mode) > 0:
            df[c] = df[c].fillna(mode[0])
        after_na = df[c].isna().sum()
        report_impute(c, after_na, before_na)

    # Booleanas
    for c in bool_cols:
        before_na = df[c].isna().sum()
        mode = df[c].mode()
        if len(mode) > 0:
            df[c] = df[c].fillna(mode[0])
        after_na = df[c].isna().sum()
        report_impute(c, after_na, before_na)

    # ==========================
    # 8) REPORTE FINAL
    # ==========================
    print("\n================ REPORTE FINAL =================")
    print("Shape final:", df.shape)

    remaining_na = df.isna().sum().sum()
    print("Total valores faltantes restantes:", remaining_na)

    if remaining_na == 0:
        print("Dataset completamente limpio üéâ")

    # ==========================
    # 9) GUARDAR
    # ==========================
    df.to_csv(OUT_CSV, index=False)
    print("\nArchivo guardado:", OUT_CSV)
    print("================ FIN DEL PROCESO =================")
```

```python
## ==========================
# IMPUTACI√ìN CORREGIDA PARA AIRBNB
# ==========================

print("\n================ IMPUTACI√ìN CORREGIDA =================")

# --------------------------
# 1) NO TOCAR COORDENADAS
# --------------------------
# latitude y longitude pueden ser negativos seg√∫n el pa√≠s
# no se imputan, si faltan ya se eliminaron en cr√≠ticas

COORD_COLS = ["latitude", "longitude"]

# --------------------------
# 2) MNAR REAL (RESE√ëAS)
# --------------------------
review_score_cols = [c for c in df.columns if c.startswith("review_scores_")]
review_like_cols = ["reviews_per_month"]
mnar_cols = review_score_cols + [c for c in review_like_cols if c in df.columns]

if mnar_cols:
    print("\nüî¥ MNAR estructural (sin reviews = 0 + flag)")
    for c in mnar_cols:
        missing_before = df[c].isna().sum()
        if missing_before > 0:
            df[f"{c}__was_missing"] = df[c].isna().astype(int)
            df[c] = df[c].fillna(0)
            print(f" - {c}: {missing_before} -> 0 + flag")

# --------------------------
# 3) VARIABLES QUE DEBEN SER >= 0
# --------------------------
non_negative_cols = [
    "price", "minimum_nights", "maximum_nights",
    "availability_30", "availability_60", "availability_90",
    "availability_365", "number_of_reviews",
    "number_of_reviews_ltm", "number_of_reviews_l30d",
    "beds", "bedrooms", "bathrooms",
    "accommodates",
    "estimated_occupancy_l365d",
    "estimated_revenue_l365d"
]

non_negative_cols = [c for c in non_negative_cols if c in df.columns]

# Convertir negativos inv√°lidos en NaN
for c in non_negative_cols:
    neg_count = (df[c] < 0).sum()
    if neg_count > 0:
        df.loc[df[c] < 0, c] = pd.NA
        print(f"üõ† {c}: {neg_count} negativos convertidos a NA")

# --------------------------
# 4) IMPUTACI√ìN NUM√âRICA INTELIGENTE
# --------------------------
GROUP_COL = "neighbourhood_cleansed"
use_group = GROUP_COL in df.columns

num_cols = df.select_dtypes(include=["number"]).columns.tolist()

# excluir:
num_cols = [
    c for c in num_cols
    if c not in mnar_cols
    and c not in COORD_COLS
]

print("\nüü¢ Imputando num√©ricas...")

if use_group:
    print(f" - Mediana por grupo ({GROUP_COL}) con fallback global")
    global_medians = df[num_cols].median(numeric_only=True)

    for c in num_cols:
        miss_before = df[c].isna().sum()
        if miss_before == 0:
            continue

        df[c] = df.groupby(GROUP_COL)[c].transform(
            lambda s: s.fillna(s.median())
        )

        df[c] = df[c].fillna(global_medians.get(c, df[c].median()))

        miss_after = df[c].isna().sum()
        print(f"   * {c}: imputados {miss_before - miss_after}")

else:
    print(" - Mediana global")
    for c in num_cols:
        miss_before = df[c].isna().sum()
        if miss_before == 0:
            continue

        df[c] = df[c].fillna(df[c].median())
        miss_after = df[c].isna().sum()
        print(f"   * {c}: imputados {miss_before - miss_after}")

# --------------------------
# 5) CLIP FINAL (SEGURIDAD)
# --------------------------
for c in non_negative_cols:
    df[c] = df[c].clip(lower=0)

# --------------------------
# 6) CATEG√ìRICAS
# --------------------------
cat_cols = df.select_dtypes(include=["object", "string", "category"]).columns

# no tocar ID ni URLs
cat_exclude = ["id", "listing_url", "host_url", "picture_url"]
cat_cols = [c for c in cat_cols if c not in cat_exclude]

print('\nüü° Imputando categ√≥ricas con "Unknown"...')

for c in cat_cols:
    miss_before = df[c].isna().sum()
    if miss_before == 0:
        continue

    df[c] = df[c].fillna("Unknown")
    miss_after = df[c].isna().sum()

    print(f"   * {c}: imputados {miss_before - miss_after}")

print("\n================ FIN IMPUTACI√ìN =================")
```

# Detectar outliers con m√©todo IQR

```python
def detectar_outliers_iqr(data, columna):
    Q1 = data[columna].quantile(0.25)
    Q3 = data[columna].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = data[(data[columna] < lower) | (data[columna] > upper)]
    
    print(f"\nüîé {columna}")
    print("Lower bound:", lower)
    print("Upper bound:", upper)
    print("Cantidad outliers:", len(outliers))
    
    return lower, upper

bounds = {}
for var in variables:
    bounds[var] = detectar_outliers_iqr(df, var)
```

```python
# ==========================
# REGRESI√ìN LINEAL DEFINITIVA (ARREGLADA)
# ==========================

import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error, mean_absolute_error

# 0) (Opcional pero recomendado) asegurar que price_log exista y sea v√°lido
y = pd.to_numeric(df["price_log"], errors="coerce")

# 1) Variables predictoras que existen
num_features_existing = [c for c in num_features_clean if c in df.columns]
print("Variables usadas:", len(num_features_existing))
print(num_features_existing)

# 2) Construcci√≥n de X
X = df[num_features_existing].copy()

# 3) LIMPIEZA ESPECIAL: columnas en porcentaje tipo "96%"
percent_cols = [c for c in ["host_response_rate", "host_acceptance_rate"] if c in X.columns]
for c in percent_cols:
    X[c] = (
        X[c].astype(str)
        .str.replace("%", "", regex=False)
        .str.strip()
        .replace("nan", np.nan)
        .replace("", np.nan)
    )
    X[c] = pd.to_numeric(X[c], errors="coerce") / 100.0  # queda 0-1

# 4) Convertir todo X a num√©rico (lo que no se pueda, queda NaN)
X = X.apply(pd.to_numeric, errors="coerce")

# 5) Limpiar inf
X = X.replace([np.inf, -np.inf], np.nan)
y = y.replace([np.inf, -np.inf], np.nan)

# 6) NO hacer dropna total: quitar SOLO filas sin y (target)
mask_y = y.notna()
X = X.loc[mask_y].copy()
y = y.loc[mask_y].copy()

# 7) Imputaci√≥n simple para X (mediana por columna)
# (evita quedarte sin filas)
X = X.fillna(X.median(numeric_only=True))

# 8) Seguridad: quitar columnas que quedaron totalmente vac√≠as (por si acaso)
all_nan_cols = X.columns[X.isna().all()].tolist()
if all_nan_cols:
    print("‚ö†Ô∏è Columnas eliminadas por quedar 100% NaN:", all_nan_cols)
    X = X.drop(columns=all_nan_cols)

# 9) Dataset final para statsmodels
X_clean = X.astype(float)
y_clean = y.astype(float)

print("Filas usadas para entrenar:", len(y_clean))
print("Shape X_clean:", X_clean.shape)

# (Opcional) Preview con datos (5 filas)
display(pd.concat([X_clean.head(5), y_clean.head(5).rename("price_log")], axis=1))

# 10) Modelo OLS
X_sm = sm.add_constant(X_clean, has_constant="add").astype(float)
modelo = sm.OLS(y_clean, X_sm).fit()

# 11) M√©tricas
y_pred = modelo.predict(X_sm)
rmse = np.sqrt(mean_squared_error(y_clean, y_pred))
mae = mean_absolute_error(y_clean, y_pred)

print("\n================ M√âTRICAS =================")
print("R¬≤:", modelo.rsquared)
print("R¬≤ Ajustado:", modelo.rsquared_adj)
print("RMSE:", rmse)
print("MAE:", mae)

print(modelo.summary())
```

```python
def diagnostic_plots(X, y, model_fit=None, top_n=3):
    """
    4 diagnostic plots like R's plot(lm):
      1) Residuals vs Fitted
      2) Normal Q-Q
      3) Scale-Location
      4) Residuals vs Leverage (Cook's distance)

    X: DataFrame/array of predictors (NO constant column needed)
    y: Series/array target
    model_fit: fitted statsmodels OLS results (optional)
    top_n: number of points to label in each plot
    """

    # --- asegurar formatos y alineaci√≥n ---
    if isinstance(X, np.ndarray):
        X = pd.DataFrame(X)
    if isinstance(y, (np.ndarray, list)):
        y = pd.Series(y)

    X = X.copy()
    y = y.copy()

    # alinear √≠ndices (clave para anotaciones)
    data = pd.concat([X, y.rename("y")], axis=1).replace([np.inf, -np.inf], np.nan).dropna()
    X = data.drop(columns=["y"])
    y = data["y"]

    # fit si no lo pasan
    if model_fit is None:
        model_fit = sm.OLS(y, sm.add_constant(X)).fit()

    infl = model_fit.get_influence()

    fitted = model_fit.fittedvalues
    resid = model_fit.resid
    stud_resid = infl.resid_studentized_internal
    leverage = infl.hat_matrix_diag
    cooks = infl.cooks_distance[0]
    sqrt_abs_stud = np.sqrt(np.abs(stud_resid))

    # --------- 1) Residuals vs Fitted ----------
    fig1, ax1 = plt.subplots(figsize=(7,5))
    sns.residplot(x=fitted, y=resid, lowess=True,
                  scatter_kws={"alpha":0.5},
                  line_kws={"lw":1, "alpha":0.8},
                  ax=ax1)
    ax1.set_title("Residuals vs Fitted")
    ax1.set_xlabel("Fitted values")
    ax1.set_ylabel("Residuals")

    # anotar top_n por |residual|
    idx_top = np.argsort(np.abs(resid))[-top_n:]
    for i in idx_top:
        ax1.annotate(str(data.index[i]), (fitted.iloc[i], resid.iloc[i]))

    # --------- 2) Normal Q-Q ----------
    fig2 = plt.figure(figsize=(7,5))
    ax2 = fig2.add_subplot(111)
    QQ = ProbPlot(stud_resid)
    QQ.qqplot(line="45", alpha=0.5, ax=ax2)
    ax2.set_title("Normal Q-Q")
    ax2.set_xlabel("Theoretical Quantiles")
    ax2.set_ylabel("Standardized Residuals")

    idx_top = np.argsort(np.abs(stud_resid))[-top_n:]
    # coordenadas para anotar en el Q-Q
    theoretical = QQ.theoretical_quantiles
    ordered = np.sort(stud_resid)
    for j, i in enumerate(idx_top):
        # buscar posici√≥n aproximada del punto i dentro de la serie ordenada
        # (esto es una aproximaci√≥n, suficiente para anotaci√≥n visual)
        pos = np.searchsorted(ordered, stud_resid[i])
        pos = np.clip(pos, 0, len(theoretical)-1)
        ax2.annotate(str(data.index[i]), (theoretical[pos], stud_resid[i]))

    # --------- 3) Scale-Location ----------
    fig3, ax3 = plt.subplots(figsize=(7,5))
    ax3.scatter(fitted, sqrt_abs_stud, alpha=0.5)
    sns.regplot(x=fitted, y=sqrt_abs_stud, scatter=False, ci=False, lowess=True,
                line_kws={"lw":1, "alpha":0.8}, ax=ax3)
    ax3.set_title("Scale-Location")
    ax3.set_xlabel("Fitted values")
    ax3.set_ylabel(r"$\sqrt{|Standardized Residuals|}$")

    idx_top = np.argsort(sqrt_abs_stud)[-top_n:]
    for i in idx_top:
        ax3.annotate(str(data.index[i]), (fitted.iloc[i], sqrt_abs_stud[i]))

    # --------- 4) Residuals vs Leverage ----------
    fig4, ax4 = plt.subplots(figsize=(7,5))
    ax4.scatter(leverage, stud_resid, alpha=0.5)
    sns.regplot(x=leverage, y=stud_resid, scatter=False, ci=False, lowess=True,
                line_kws={"lw":1, "alpha":0.8}, ax=ax4)
    ax4.set_title("Residuals vs Leverage")
    ax4.set_xlabel("Leverage")
    ax4.set_ylabel("Standardized Residuals")
    ax4.set_xlim(0, leverage.max() + 0.01)

    # anotar top_n por Cook's distance
    idx_top = np.argsort(cooks)[-top_n:]
    for i in idx_top:
        ax4.annotate(str(data.index[i]), (leverage[i], stud_resid[i]))

    # l√≠neas de Cook (0.5 y 1)
    p = int(model_fit.df_model) + 1  # params aprox (incluye intercepto)
    lev_space = np.linspace(0.001, leverage.max(), 100)
    cook_05 = np.sqrt((0.5 * p * (1 - lev_space)) / lev_space)
    cook_1 = np.sqrt((1.0 * p * (1 - lev_space)) / lev_space)

    ax4.plot(lev_space, cook_05, ls="--")
    ax4.plot(lev_space, -cook_05, ls="--")
    ax4.plot(lev_space, cook_1, ls=":")
    ax4.plot(lev_space, -cook_1, ls=":")

    plt.show()

    return model_fit
```
